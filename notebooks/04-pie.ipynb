# Pie

```python
import matplotlib.pyplot as plt

# Data to plot
labels = ['LOC, BRANCH_C & UNIQ_OPER',
'ESS_COMP, LOC_EXE & UNIQ_OPER',
'LOC, LOC_EXE & ESSENT_COMP',
'LOC_EXE, BRAN_C & ESS_COMP',
'BRAN_C, ESS_COMP & LOC_EXE']
sizes = [25,7,3,3,2]
colors = ['dimgray', 'grey', 'darkgray', 'lightgrey', 'whitesmoke']
```

```python
# Plot
plt.pie(sizes, 
        labels=labels, 
        colors=colors, 
        autopct='%1.1f%%', 
        shadow=True, 
        startangle=140)

plt.axis('equal')
plt.savefig('../reports/q1.eps', bbox_inches='tight')
plt.show()
```
